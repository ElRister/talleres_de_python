```python
def obtener_numero_matrices():
    while True:
        num_matrices = int(input("Ingrese la cantidad de matrices que desea sumar: "))
        if num_matrices >= 2:
            return num_matrices
        else:
            print("Debe ingresar al menos 2 matrices para realizar la suma.")

def obtener_dimensiones_matrices():
    num_filas = int(input("Ingrese el número de filas de las matrices: "))
    num_columnas = int(input("Ingrese el número de columnas de las matrices: "))
    return num_filas, num_columnas

def obtener_matrices(num_matrices, num_filas, num_columnas):
    matrices = []
    for i in range(num_matrices):
        matriz = []
        print(f"Ingrese los elementos de la matriz {i + 1}:")
        for fila in range(num_filas):
            fila_matriz = []
            for columna in range(num_columnas):
                valor = int(input(f"Ingrese el valor para la posición [{fila + 1}][{columna + 1}]: "))
                fila_matriz.append(valor)
            matriz.append(fila_matriz)
        matrices.append(matriz)
    return matrices

def sumar_matrices(matrices):
    if len(matrices) < 2:
        return None
    
    num_filas = len(matrices[0])
    num_columnas = len(matrices[0][0])
    
    for matriz in matrices:
        if len(matriz) != num_filas or len(matriz[0]) != num_columnas:
            return None
    
    suma = [[0] * num_columnas for _ in range(num_filas)]
    
    for matriz in matrices:
        for i in range(num_filas):
            for j in range(num_columnas):
                suma[i][j] += matriz[i][j]
    
    return suma

def mostrar_matrices(matrices):
    print("Matrices:")
    for i, matriz in enumerate(matrices):
        print(f"Matriz {i + 1}:")
        for fila in matriz:
            print(fila)
    print("\nMatriz Resultante:")
    if matrices[-1] is not None:
        for fila in matrices[-1]:
            print(fila)
    else:
        print("No se pueden sumar las matrices debido a diferencias en las dimensiones.")

if __name__ == "__main__":
    num_matrices = obtener_numero_matrices()
    num_filas, num_columnas = obtener_dimensiones_matrices()
    matrices = obtener_matrices(num_matrices, num_filas, num_columnas)
    resultado = sumar_matrices(matrices)
    matrices.append(resultado)
    mostrar_matrices(matrices)
```
